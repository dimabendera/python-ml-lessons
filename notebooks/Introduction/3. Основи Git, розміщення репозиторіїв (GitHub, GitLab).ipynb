<h1>3. Основи Git, розміщення репозиторіїв (GitHub, GitLab)</h1>

<p>Git і GitHub є поширеними інструментами, які використовуються в програмуванні. Вони допомагають вам керувати різними версіями вашого коду та співпрацювати з іншими розробниками.</p>

<h2>Що таке Git і GitHub?</h2>

<p>Git був розроблений у 2005 році Лінусом Торвальдсом як програмне забезпечення з відкритим кодом для відстеження змін у розподіленій системі керування версіями .</p>

<p>Git є відкритим вихідним кодом, оскільки його вихідний код надається у вільний доступ для будь-кого для зміни та використання. Проекти з відкритим кодом створюються та підтримуються спільно різними розробниками в різних місцях.</p>

<p>Git відстежує зміни через розподілену систему керування версіями. Це означає, що Git може відстежувати стан різних версій ваших проектів під час їх розробки. Він розповсюджується, тому що ви можете отримати доступ до своїх кодових файлів з іншого комп’ютера, а також інші розробники.</p>

<p>Коли ви створюєте проект з відкритим вихідним кодом, вам знадобиться спосіб документувати або відстежувати свій код. Це допоможе впорядкувати вашу роботу та відстежувати внесені зміни. Це те, що Git дозволяє вам робити.</p>


<p>Але вам також потрібне місце для розміщення вашого коду, що спрощує та пришвидшує керування кожною версією вашого проекту. Тут на допомогу приходить GitHub.</p>

<p>GitHub — це «хаб» (місце або платформа), де користувачі Git разом створюють програмне забезпечення. GitHub також є хостинг-провайдером і платформою контролю версій, яку можна використовувати для спільної роботи над проектами з відкритим кодом і обміну файлами. Коли ви використовуєте GitHub, ви працюєте з Git під капотом.</p>

<p>Отже, Git — це фактична система контролю версій, а GitHub — платформа, на якій ви розміщуєте свій код.</p>


<h2>Як почати використовувати Git і GitHub</h2>
<h4>Крок 1 – Встановіть Git</h4>
<p>Git попередньо встановлено на деяких комп’ютерах Mac і Linux, але ви завжди можете перевірити, чи встановлено Git на вашому комп’ютері, ввівши <code>git version</code> в термінал. Для цього можна скористатися командним рядком.</p>

<img src="./assets/checkGItInstalled-1.png"></img>

<p>Якщо в вас не встановлений git, ви можете завантажити його <a href="https://git-scm.com/download">тут</a>, а потім вибрати операційну систему для завантаження.</p>

<img src="./assets/gitDownload-1.png"></img>

<p>Дотримуйтеся вказівок необхідного посібника для встановлення, доки встановлення не завершиться. Відкрийте командний рядок і введіть <code>git version</code>, щоб переконатися, що Git успішно встановлено.</p>


<h4>Крок 2 – Створення облікового запису GitHub.</h4>

Створіть обліковий запис на <a href="https://github.com/">GitHub.com</a>.

<img src="./assets/githubDownload-1.png"></img>

<p>Щоб створити обліковий запис на GitHub, вас попросять надати деяку особисту інформацію, як-от ім’я, підтвердити вашу електронну адресу, встановити ім’я користувача та пароль, і ваш обліковий запис буде налаштовано за лічені хвилини.</p>



<h4>Крок 3. Підключіть обліковий запис GitHub до облікового запису Git.</h4>
<p>Щоб установити своє ім’я користувача Git, введіть це у своєму терміналі:</p>
<pre><code>git config --global user.name "Example Username"</code></pre>

<p>Щоб підтвердити, що ви правильно встановили ім’я користувача Git, введіть це:</p>
<pre><code>git config --global user.name</code></pre>
<p>Ви маєте отримати "Example Username" як результат.</p>

<p>Щоб налаштувати електронну пошту Git, введіть у своєму терміналі:</p>
<pre><code>git config --global user.email "youremail@gmail.com"</code></pre>

<p>Щоб підтвердити, що ви правильно налаштували електронну пошту Git, введіть це:</p>
<pre><code>git config --global user.email</code></pre>
<p>У вихідних даних має бути "youremail@gmail.com".</p>
<p>Вам буде запропоновано автентифікувати обліковий запис GitHub, тож просто увійдіть, використовуючи ту саму електронну адресу, щоб підтвердити.</p>
