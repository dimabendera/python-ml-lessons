<h4>Бізнес-задача</h4>
<p>Більшість проєктів починається з бізнес-задачі, формулювання якої далеке від Data Science задачі. Головне завдання на цьому етапі — зрозуміти потреби бізнесу і транслювати їх у задачу Data Science. На цьому етапі варто поставити декілька питань:</p>

<ol> 
  <li type="-">Чи дійсно задачу найефективніше вирішувати засобами Data Science/ Machine Learning? У деяких випадках відповідь буде «ні», і вчасно поставлене питання збереже час і ресурси бізнес-стейкхолдерів і Data Science-команди.</li>
  <li type="-">Як ми будемо використовувати Machine Learning модель, коли отримаємо її на практиці? Чим чіткішою буде відповідь на це питання на початку проєкту, тим кращим буде ML-рішення. Відповідь на це питання часто впливає на вибір DS/ML-рішення.</li>
  <li type="-">Чи є у нас дані для побудови ML/DS рішення? Якщо даних дуже мало, можливо, спочатку потрібно налаштувати збір даних.</li>
  <li type="-">Чи готові ми технічно до використання ML-моделі? Бувають випадки, коли ML-модель принесла вигоду бізнесу, для запуску моделі достатньо даних, але технічна інфраструктура потребує змін, щоб інтегрувати результати моделі у процес прийняття рішень.</li>
</ol>
<p>Якщо відповідь на усі питання позитивна — можна переходити до наступних етапів.</p>

<h4>Prototype</h4>
<p>Якщо після аналізу поставленої бізнес-задачі вирішили працювати над моделлю, час переходити до створення прототипу моделі. Цей етап складається з декількох кроків, розглянемо кожен з них окремо.</p>

<h4>Data extraction</h4>

<p>Один із найважливіших кроків — побудова початкового датасету, що ляже в основу першої версії моделі. Алгоритм дії на цьому етапі наступний:</p>

<ul> 
  <li>Оцінити, які є дані. Зокрема, зрозуміти, чи достатньо історичних даних для побудови цільової змінної.</li>
  <li>Розбити дані на декілька груп залежно від доступності і необхідності попередньої обробки перед використанням. Наприклад, в першу чергу варто подивитися на дані, які не потребують попередньої обробки перед використанням в моделі і можуть бути отримані швидко. Із таких даних можна побудувати першу версію датасету. Далі потрібно виділити групу даних, що потребує посередніх зусиль і часу для обробки. Залежно від корисності та об’єму даних з початкової групи, працювати з ними одразу або відкласти на наступну ітерацію. Дані, які складно дістати, або потрібно довго чистити чи агрегувати, варто відкласти до наступних ітерацій.</li>
  <li>Побудувати першу версії датасету, на основі якої можна будувати модель.</li>
</ul>

<h4>MVP</h4>
<p>Найцікавіший момент для кожного Data Scientist — створення першої версії моделі. Проте, перш ніж переходити до найбільш інтригуючого моменту, варто не пропустити кілька критично важливих кроків.</p>


<p><b>Остаточне визначення цільової змінної.</b>Цей крок належить саме до цього етапу, оскільки потребує аналізу і обговорення з бізнес-стейкхолдерами. Зазвичай навіть у найбільш очевидних випадках цільова має декілька варіантів.</p>
<p><b>Трансформація даних.</b>Прибирання викидів, обробка пропущених значень, кодування категоріальних змінних мають критичний вплив на точність і стабільність моделі.</p>
<p><b>Стратифікація даних.</b>Якщо даних дуже багато, потрібно коректно сформувати репрезентативну вибірку для моделювання.</p>
<p><b>Вибір алгоритму.</b>Варто вибирати алгоритм, що найкраще підходить під конкретну задачу.</p>
<p><b>Власне моделювання.</b>Створення першої версії моделі, яку ми потім використовуватимемо як базу для порівняння. Усі наступні версії мають бути кращими за базову. Також перша версія допоможе зрозуміти, чи можна створити модель на першому наборі даних, чи потрібно повертатися до етапу Data extraction і діставати додаткові дані.</p>

<h4>Валідація моделі</h4>
<p>Додаткова верифікація моделі на ранніх етапах розробки допомагає уникнути багатьох помилок і отримати цінні ідеї для швидкого покращення. Варто розглянути варіант верифікації моделі як іншими Data Scientists, так і бізнес-стейкхолдерами. Для цього можна обрати різні механізми.</p>

<p>Для додаткової перевірки моделі колегами, можна поширити прототип моделі через git або у ноутбуці через Databricks або Google Colab.</p>

<p>Для валідації моделі бізнес-стейкхолдерами існує багато різних механізмів, наприклад, узагальнити результати моделі для певної підвибірки і показати, наскільки гарно модель прогнозує чи розділяє дані. Або порахувати результати моделі для підвибірки даних і віддати на валідацію стейкхолдерам.</p>

<h4>Деплоймент нової моделі</h4>
<p>Якщо етап валідації пройдено вдало, час вирішувати питання деплойменту. Тут є декілька важливих питань.</p>

<ul> 
  <li>Наскільки часто потрібно оновлювати прогноз моделі? Від відповіді на це питання залежить вибір механізму деплою.
      <ul> 
          <li>Якщо дані потрібні в реальному часі, потрібно буде створювати мікросервіс.</li>
          <li>Якщо дані достатньо оновлювати декілька разів на день або рідше, можна зробити батчеву імплементацію, коли модель буде запускатися, збирати необхідні дані, рахувати прогноз і складати отримані дані у заздалегідь узгоджене місце (Data Lake, Database etc).</li>
      </ul>
  </li>
  <li>Чи потрібно автоматично перетреновувати модель? Якщо так, тоді потрібно одразу створити окремий процес для тренування моделі.</li>
</ul>


<h4>Впровадження зворотного зв’язку</h4>
<p>Протягом перших тижнів/місяців роботи моделі у реальному середовищі починають накопичуватись ідеї для подальшого покращення. Ці ідеї пов’язані з:<p>
<ul> 
  <li>виявленням недоліків прогнозування для певних сегментів популяції;</li>
  <li>появою нових перспективних даних, що можуть покращити результати моделі;</li>
  <li>усвідомленням бізнесу, що невелике покращення точності може мати суттєвий позитивний вплив на бізнес-метрики;</li>
  <li>необхідністю змінити чи підкоригувати цільову змінну, якщо вона не повністю відповідає бізнес-задачам;</li>
  <li>зміною бізнес-процесів і необхідністю адаптувати модель до нових реалій.</li>
</ul>
<p>Якщо немає термінових запитів на зміну, варто накопичувати такі зміни і планово робити перегляд логіки моделі. Періодичність дуже сильно залежатиме від динаміки бізнесу і конкретної задачі.</p>

<h2>Як працювати з командою розробки продукту: найкращі практики?</h2>

<p>Робота з командою розробки продуктів вимагає ефективного менеджменту та комунікації для досягнення успішних результатів. Ось деякі практичні поради, які допоможуть вам співпрацювати з командою розробників продукту.</p>
<ol> 
  <li type="1">Встановіть чіткі цілі та очікування: визначте чіткі цілі та очікування щодо функціональності, часу та якості продукту. Обговоріть і уточніть їх із командою, щоб усі учасники розуміли, чого від них очікують.</li>
  <li type="1">Встановіть ефективні канали спілкування: забезпечте регулярне та відкрите спілкування з командою. Використовуйте інструменти спілкування, такі як зустрічі, відеоконференції, чати або дошки для спільної роботи, щоб налагодити ефективні канали спілкування та обміну інформацією.</li>
  <li type="1">Призначте відповідальність і делегуйте завдання: визначте ролі та обов’язки в команді, враховуючи навички та досвід кожного члена. Призначайте завдання та делегуйте їх, враховуючи сильні сторони кожного члена команди.</li>
    <li type="1">Зберігайте прозорість і відкритість: прагніть до відкритого обміну інформацією та ідеями. Зберігайте прозорість щодо прогресу роботи, проблем і проблем. Сприяйте створенню атмосфери, в якій кожному члену команди буде комфортно ділитися своїми думками та пропонувати ідеї.</li>
    <li type="1">Вчіться на прикладах найкращих практик: вивчайте успішні приклади роботи з командами розробки продуктів, шукайте досвід і уроки, які можна застосувати у вашому проекті. Діліться знаннями та досвідом з іншими менеджерами з продуктів або беріть участь у професійних спільнотах, щоб отримати цінні ідеї.</li>
    <li type="1">Оцінка та відгук: регулярно надавайте відгук членам команди. Оцінюйте та визнавайте їхні досягнення, а також пропонуйте конструктивну критику для покращення їхньої роботи. Враховуйте думки та пропозиції колективу та застосовуйте їх у своїй роботі.</li>
</ol>

<h2>Методи управління командою розробки</h2>

<p>Управління командою розробки передбачає використання різноманітних методів і підходів для координації роботи, мотивації членів команди, планування та контролю процесу розробки. Ось кілька методів, які можна використовувати під час керування командою розробників, із прикладами та детальним описом:</p>

<h4>Гнучкі методології</h4>

<p>Scrum: методологія, заснована на ітераційній розробці та поступовій доставці продукту. Команда ділить роботу на короткі сегменти часу, які називаються спринтами, і проводить регулярні зустрічі, відомі як «Щоденний стенд», «Планування спринту», «Огляд спринту» та «Ретроспектива спринту».</p>

<p>Канбан: методологія, у якій робота представлена ​​у формі карток, що переміщуються по дошці від етапу до етапу. Команда візуально відстежує хід роботи та керує потоком завдань.</p>

<h4>DevOps</h4>
<p>Безперервна інтеграція (CI) і безперервна доставка (CD): методи, спрямовані на автоматизацію процесу інтеграції, тестування та доставки коду в розробці програмного забезпечення. Команда використовує такі інструменти, як Jenkins або GitLab CI/CD, для автоматичного створення, тестування та розгортання програм.</p>
<p>Інфраструктура як код (IaC): підхід, у якому інфраструктура для розробки та розгортання програм описується у формі коду. Команда використовує такі інструменти, як Terraform або Ansible, для автоматизації розгортання та управління інфраструктурою.</p>

<h4>Бережливий підхід</h4>
<p>Відображення потоку цінностей: метод аналізу потоку роботи в команді розробників, виявлення вузьких місць і оптимізації процесів. Команда аналізує поточний робочий процес, виявляє неефективність і впроваджує вдосконалення для підвищення ефективності та якості розробки.</p>
<p>Кайдзен: підхід, заснований на постійному вдосконаленні процесу. Команда постійно прагне до досконалості, щодня вносить невеликі покращення та заохочує ініціативу всіх учасників.</p>

<h4>Гнучкі практики</h4>
<p>Історії користувачів: методологія, у якій вимоги до продукту представлені у формі коротких описів сценаріїв використання з боку користувача. Команда розробників використовує Історії користувачів, щоб зрозуміти потреби користувачів і визначити функціональність, яку потрібно розробити.</p>
<p>Ітеративна розробка: підхід, при якому розробка продукту відбувається поетапно з поступовим додаванням ітерацій і функціональності. Команда випускає версії продукту на кожному етапі розробки та збирає відгуки користувачів, щоб внести корективи та покращити продукт</p>

<h4>Розподіл завдань</h4>
<p>Спліт-тестування (тестування A/B): метод, за якого команда ділить аудиторію на дві або більше груп і пропонує їм різні варіанти продукту чи функціональності. Потім продуктивність варіантів порівнюється, щоб вирішити, яку версію використовувати.</p>
<p>Розподіл ролей: команда розробників може розділяти завдання на менші та призначати їх різним спеціалістам у команді. Наприклад, розділіть завдання на розробку інтерфейсу, розробку бекенда, тестування та дизайн.</p>

<h4>Комунікація та зворотній зв'язок</h4>
<p>Щоденні стендап-зустрічі: щоденні зустрічі, на яких члени команди діляться інформацією про свою роботу, прогрес, проблеми та плани на день. Це дозволяє всім учасникам бути в курсі поточного стану проекту та спільно вирішувати проблеми.</p>
<p>Ретроспективні зустрічі: регулярні зустрічі, на яких команда аналізує минулий період роботи, визначає вдосконалення та обговорює, як зробити процес розробки ефективнішим.</p>
