<h1>3. Основи Git, розміщення репозиторіїв (GitHub, GitLab)</h1>

<p>Git і GitHub є поширеними інструментами, які використовуються в програмуванні. Вони допомагають вам керувати різними версіями вашого коду та співпрацювати з іншими розробниками.</p>

<h2>Що таке Git і GitHub?</h2>

<p>Git був розроблений у 2005 році Лінусом Торвальдсом як програмне забезпечення з відкритим кодом для відстеження змін у розподіленій системі керування версіями .</p>

<p>Git є відкритим вихідним кодом, оскільки його вихідний код надається у вільний доступ для будь-кого для зміни та використання. Проекти з відкритим кодом створюються та підтримуються спільно різними розробниками в різних місцях.</p>

<p>Git відстежує зміни через розподілену систему керування версіями. Це означає, що Git може відстежувати стан різних версій ваших проектів під час їх розробки. Він розповсюджується, тому що ви можете отримати доступ до своїх кодових файлів з іншого комп’ютера, а також інші розробники.</p>

<p>Коли ви створюєте проект з відкритим вихідним кодом, вам знадобиться спосіб документувати або відстежувати свій код. Це допоможе впорядкувати вашу роботу та відстежувати внесені зміни. Це те, що Git дозволяє вам робити.</p>


<p>Але вам також потрібне місце для розміщення вашого коду, що спрощує та пришвидшує керування кожною версією вашого проекту. Тут на допомогу приходить GitHub.</p>

<p>GitHub — це «хаб» (місце або платформа), де користувачі Git разом створюють програмне забезпечення. GitHub також є хостинг-провайдером і платформою контролю версій, яку можна використовувати для спільної роботи над проектами з відкритим кодом і обміну файлами. Коли ви використовуєте GitHub, ви працюєте з Git під капотом.</p>

<p>Отже, Git — це фактична система контролю версій, а GitHub — платформа, на якій ви розміщуєте свій код.</p>


<h2>Як почати використовувати Git і GitHub</h2>
<h4>Крок 1 – Встановіть Git</h4>
<p>Git попередньо встановлено на деяких комп’ютерах Mac і Linux, але ви завжди можете перевірити, чи встановлено Git на вашому комп’ютері, ввівши <code>git version</code> в термінал. Для цього можна скористатися командним рядком.</p>

<img src="./assets/checkGItInstalled-1.png"></img>

<p>Якщо в вас не встановлений git, ви можете завантажити його <a href="https://git-scm.com/download">тут</a>, а потім вибрати операційну систему для завантаження.</p>

<img src="./assets/gitDownload-1.png"></img>

<p>Дотримуйтеся вказівок необхідного посібника для встановлення, доки встановлення не завершиться. Відкрийте командний рядок і введіть <code>git version</code>, щоб переконатися, що Git успішно встановлено.</p>


<h4>Крок 2 – Створення облікового запису GitHub.</h4>

Створіть обліковий запис на <a href="https://github.com/">GitHub.com</a>.

<img src="./assets/githubDownload-1.png"></img>

<p>Щоб створити обліковий запис на GitHub, вас попросять надати деяку особисту інформацію, як-от ім’я, підтвердити вашу електронну адресу, встановити ім’я користувача та пароль, і ваш обліковий запис буде налаштовано за лічені хвилини.</p>



<h4>Крок 3. Підключіть обліковий запис GitHub до облікового запису Git.</h4>
<p>Щоб установити своє ім’я користувача Git, введіть це у своєму терміналі:</p>
<pre><code>git config --global user.name "Example Username"</code></pre>

<p>Щоб підтвердити, що ви правильно встановили ім’я користувача Git, введіть це:</p>
<pre><code>git config --global user.name</code></pre>
<p>Ви маєте отримати "Example Username" як результат.</p>

<p>Щоб налаштувати електронну пошту Git, введіть у своєму терміналі:</p>
<pre><code>git config --global user.email "youremail@gmail.com"</code></pre>

<p>Щоб підтвердити, що ви правильно налаштували електронну пошту Git, введіть це:</p>
<pre><code>git config --global user.email</code></pre>
<p>У вихідних даних має бути "youremail@gmail.com".</p>
<p>Вам буде запропоновано автентифікувати обліковий запис GitHub, тож просто увійдіть, використовуючи ту саму електронну адресу, щоб підтвердити.</p>

<p>Натисніть кнопку «Створити репозиторій», щоб створити репозиторій. Вам буде запропоновано перейти на цю сторінку:</p>

<h4>Крок 6. Надішліть свій локальний код на GitHub</h4>
<p>Ви можете використовувати вбудований термінал редактора коду, щоб використовувати Git для надсилання коду на GitHub. Натисніть <code>ctrl+shift+'</code>, щоб відкрити термінал у VSCode або відерийте gitbash в локальній папці вашого проету.</p>

<p>Введіть наведені нижче команди одну за одною у свій термінал. Натисніть <code>Enter</code> клавішу для продовження після кожного введення.</p>

<pre><code>
echo "# repo_name" >> README.md
git init
git add .
git commit -m "first commit"
git branch -M main
git remote add origin https://github.com/dimabendera/repo_name.git
git push -u origin main
</code></pre>

<p>Зверніть увагу, що ми бачили <code>git add README.md</code> в репозиторії на GitHub. Але тут вводимо <code>git add .</code> можливість дозволити Git додати всі файли нашого коду замість файлу, <code>README.md</code> який створить <code>echo "# repo_name" >> README.md</code>. Отже, якщо ви створили інші файли у своїй локальній папці, вам потрібно використовувати <code>git add .</code>, щоб додати всі файли.</p>

<p>Зверніть увагу, що команда <code>git remote add origin https://github.com/dimabendera/repo_name.git</code> має містити посилання на ваше власне сховище та назву вашого облікового запису GitHub.</p>


<p><code>git init</code> Створити порожнє сховище Git або повторно ініціалізуйте існуюче</p>
<p><code>git add Readme.md</code> Додати вміст файлу до індексу </p>
<p><code>git commit -m "Added Readme.md"</code> зберігає зміни доданих до індексу файлів до локального репозиторія</p>
<p><code>git branch</code> Вивести список, створити або видалити гілку</p>
<p><code>git remote add origin https://github.com/dimabendera/repo_name.git</code>підключає локальну папку до репозиторію на GitHub.</p>
<p><code>git push -u origin main</code>надсилає код на GitHub. Прапорець <code>-u</code> створює посилання відстеження для гілки та <code>origin main</code> розміщує код у <code>main</code> гілці.</p>

<h4>Додати зміни до репозиторію</h4>
<p>Створимо новий файл <code>todo.txt</code> з вмістом:</p>
<pre><code>
MY TO-DO LIST

1. Write an article.
2. Code.
3. Study books.
4. Practice code
</code></pre>

<p>Виконаємо команду <code>git status</code>. Вона покаже зміни в нащому локальному репозиторії.</p>
<p>Оскільки новий файл ще не індексуеться додаєио його до індексу <code>git add todo.txt</code></p>
<p>Потім ми фіксуємо зміни, запустивши, <code>git commit -m "added new task"</code> а потім відсилаємо зміни на GitHub репозиторій <code>git push -u origin main</code>.</p>


<h4>Гілки в Git</h4>
<p>За допомогою гілок ви можете створити копію файлу, над яким хочете працювати, не зіпсувавши оригінальну копію. Ви можете об’єднати ці зміни з оригінальною копією або просто дозволити гілці залишатися незалежною.</p>
<p>Перш ніж перейти до використання гілок, я хочу показати вам візуальне представлення нашого репо:</p>
<img src="./assets/g638.png"></img>
<p>На зображенні вище показано нашу головну гілку з останніми двома комітами (першим комітом і доданим новим комітом завдання).</p>

<p>На даний момент я хочу додати більше завдань до списку, але я ще не впевнений, чи хочу я їх у своєму основному списку. Тож я створю нову гілку під назвою test, щоб побачити, як виглядатиме мій список із додаванням додаткових завдань.</p>
<p>Щоб створити нову гілку, виконайте цю команду: <code>git checkout -b test</code>.</p>
<p><code>checkout</code> повідомляє Git, що має перейти до нової гілки. <code>-b</code> повідомляє Git створити нову гілку. <code>test</code> це назва гілки, яку потрібно створити та переключитиcь на неї.</p>
<img src="./assets/g664.png"></img>

<p>Ми створили нову гілку зі стану нашого останнього коміту. Давайте тепер додамо більше завдань до цієї нової гілки.</p>
<pre><code>
MY TO-DO LIST

1. Write an article.
2. Code.
3. Study books.
4. Attend classes on time.
5. Practice code
6. Check exam schedule. 
</code></pre>

<p>Додайте файл, а потім зафіксуйте його зміни (зверніться до попереднього як це зробити).</p>
<p>Після фіксації змін тестової гілки поверніться до основної гілки, виконавши команду: <code>git checkout main</code>.</p>
<p>З флагом <code>-b</code> ми не створюємо нову гілку, а переходимо до існуючої. Ви можете перевірити всі гілки, які існують у вашому репо, виконавши команду <code>git branch</code>.
<p>Тепер ми можемо об’єднати зміни, зроблені в тестовій гілці, в основну гілку, запустивши <code>git merge test</code>. </p>
