<h2>Інструкції щодо встановлення Python</h2>
<p>Перейдіть на сторінку <code>https://www.python.org/downloads/windows/</code> і знайдіть стабільну версію Python 3. Натисніть відповідне посилання для вашої системи, щоб завантажити виконуваний файл: Windows installer (64-bit) або Windows installer (32-bit)</p>
<p>Python сумісний з Windows, macOS, Linux (На багатьох дистрибутивах Linux він йде разом з операційною системою). Якщо вам потрібна певна версія Python, ви також можете знайти список усіх попередніх випусків на сторінці завантаження внизу.</p>

<p>Запустіть скачаний файл <code>.exe</code> від імені адміністратора.</p>

<p>Виберіть Встановити для всіх користувачів та установіть прапорець "Add python.exe to PATH", який дає змогу користувачам запускати Python із командного рядка.</p>
<p>Якщо ви тільки починаєте працювати з Python і бажаєте встановити його з функціями за замовчуванням оберіть «Install now» і натисніть Next.</p>
<img src="./assets/CONTINT-1526_py-installer-customize.png"></img>    

  
<p>Відобразиться діалогове вікно додаткових параметрів</p>

<p>Виберіть варіанти, які відповідають вашим вимогам:</p>
<ol> 
    <li type="1"><b>Install for all users</b> рекомендується, якщо ви не єдиний користувач цього комп’ютера</li>
    <li type="1"><b>Associate files with Python</b> рекомендовано, оскільки цей параметр пов’язує всі типи файлів Python із програмою запуску чи редактором</li>

<h4>Jupyter Notebook</h4>
<p>Тут ми розглянемо як встановити <a src="https://jupyter.org/">jupyter notebook</a> і як їм користуватись</p>

<img src="./assets/jupyterlab.webp"></img>

<p><b>Jupyter Notebook</b> (раніше IPython Notebooks) — це серверно-клієнтська програма, яка дозволяє редагувати та запускати документи блокнота через веб-браузер. Програму Jupyter Notebook можна запустити на локальному робочому столі, не потребуючи доступу до Інтернету, або її можна встановити на віддаленому сервері та отримати доступ через Інтернет.</p>

<p><b>Документи блокнотів</b> (або «блокноти», всі малі регістри) — це документи, створені програмою Jupyter Notebook App , які містять як комп’ютерний код (наприклад, Python), так і елементи форматованого тексту (абзац, рівняння, малюнки, посилання тощо). Документи блокнота – це як документи, які можна прочитати людиною, що містять опис аналізу та результати (малюнки, таблиці тощо), а також виконувані документи, які можна запускати для виконання аналізу даних.</p>

<img src="./assets/229563554-2776b489-a831-44a0-a7a2-2f211e38b78b.gif"></img>

<p><b>Ядро блокнота</b> — це «обчислювальний механізм», який виконує код, що міститься в документі блокнота. Ядро ipython, виконує код python. Існують ядра для багатьох інших мов.</p>


<p>Коли ви відкриваєте документ Блокнота, автоматично запускається відповідне ядро. Коли блокнот виконується (комірка за коміркою або за допомогою меню « Комірка» -> «Виконати все» ), ядро виконує обчислення та видає результати. Залежно від типу обчислень, ядро може споживати значну кількість процесора та оперативної пам’яті. Зауважте, що оперативна пам’ять не звільняється, доки ядро не буде завершено.</p>

<p><b>Інформаційна панель Notebook</b> – це компонент, який відображається першим, коли ви запускаєте програму Jupyter Notebook App . Інформаційна панель блокнота в основному використовується для відкриття документів блокнота та керування запущеними ядрами (візуалізація та завершення роботи).</p>

<p>Інформаційна панель блокнота має інші функції, схожі на файловий менеджер, а саме навігацію папками та перейменування/видалення файлів.</p>

<h4>Як встановити jupyter?</h4>
<p>«Нульовий» крок полягає в установці сучасного стандартного браузера. Mozilla Firefox або Google Chrome працюватимуть добре. Намагайтеся уникати MS Explorer.</p>

<p>ви можете встановити Jupyter за допомогою менеджера пакетів Python - pip</p>
<p>По-перше, переконайтеся, що у вас є останній піп; старіші версії можуть мати проблеми з деякими залежностями:</p>
<pre><code>pip3 install --upgrade pip</code></pre>
<p>Потім встановіть Jupyter Notebook за допомогою:</p>
<pre><code>pip3 install jupyter</code></pre>

<h4>Запуск Jupyter Notebook</h4>
<p>Програму Jupyter Notebook можна запустити ввівши в терміналі (cmd у Windows) в робочій дерикторії:</p>
<pre><code>jupyter notebook</code></pre>

<p>Відкриється нова вкладка в вашому дефолтному браузері (якщо не відкрилась в браузері відкрийте посилання <a href="http://localhost:8888">http://localhost:8888</a>), на якій буде показано інформаційну панель блокнота</p>
<p>Після запуску програма Jupyter Notebook може отримати доступ лише до файлів у своїй папці запуску (включаючи будь-яку підпапку).</p>
<p>Закриття браузера (або вкладки) не закриє програму Jupyter Notebook . Щоб повністю вимкнути його, потрібно закрити відповідний термінал .</p>

<h4>Робота з документом</h4>
<img src="./assets/dashboard_files_tab.png"></img>

<p>На інформаційній панелі блокнота знайдіть блокнот: клацніть його назву, щоб відкрити його в новій вкладці браузера або створіть новий документ.</p>
<img src="./assets/dashboard_files_tab_new.png"></img>

<p>Натисніть меню Довідка -> Огляд інтерфейсу користувача, щоб переглянути огляд інтерфейсу користувача Jupyter Notebook App .</p>
<p>Ви можете запустити документ блокнота крок за кроком (по одній клітинці за раз), натиснувши Shift + Enter .</p>
<p>Ви можете запустити весь блокнот за один крок, натиснувши меню « Клітинка» -> «Запустити все» .</p>
<p>Щоб перезапустити ядро, клацніть у меню Ядро -> Перезапустити . Це може бути корисним, щоб почати обчислення з нуля.</p>
<img src="./assets/jupyter_ui.png"></img>

<p>Зберегти блокноти : зміни в блокнотах автоматично зберігаються кожні кілька хвилин. Щоб уникнути змін оригінального блокнота, зробіть копію документа блокнота (меню « Файл» -> «Створити копію… ») і збережіть зміни на копії.</p>
