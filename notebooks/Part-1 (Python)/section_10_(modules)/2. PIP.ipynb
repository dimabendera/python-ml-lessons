pip - це система керування пакетами для Python, яка дозволяє вам встановлювати та керувати додатковими бібліотеками та залежностями, які не входять до стандартної бібліотеки Python.

## Основні команди pip

Встановлення пакетів

````python
!pip install package-name
````

Видалення пакетів

````python
!pip uninstall package-name
````

Оновлення пакетів

````python
!pip install --upgrade package-name
````

Перегляд встановлених пакетів

````python
!pip list
````

Перевірка доступних оновлень для пакетів

````python
!pip list --outdated
````

Збереження встановлених пакетів у файл

````python
!pip freeze > requirements.txt
````

Встановлення пакетів з файлу

````python
!pip install -r requirements.txt
````

### Формат requirements.txt

Коментарі:
Ви можете додавати коментарі у requirements.txt, використовуючи символ `#`:

```
flask==1.1.2  # Основний веб-фреймворк
```

Простий список пакетів:

```
flask
requests
numpy
```

Вказівка конкретних версій:
* `==:` точна версія
* `>=:` мінімальна версія
* `<=:` максимальна версія
* `!=:` виключити конкретну версію
* `~=:` відповідає діапазону версій. Наприклад, `~=3.0` відповідає версіям від 3.0 до 4.0, але не включаючи 4.0.

```
flask==1.1.2
requests>=2.20.0
numpy<=1.18.0
```

Діапазони версій:

```
Django>=3.0,<4.0
```

Залежності з інших джерел:

Іноді може бути потрібно встановити пакет безпосередньо з репозиторію (наприклад, GitHub). Це можна зробити так:

```
# Пряме встановлення через pip:
package_name @ git+https://github.com/username/repository.git#egg=package_name

# Вказівка конкретної гілки:
package_name @ git+https://github.com/username/repository.git@branch_name

# Встановлення конкретного коміту:
package_name @ git+https://github.com/username/repository.git@commit_hash

# Вказівка піддиректорії:
package_name @ git+https://github.com/username/repository.git#subdirectory=package_dir
```

Залежності з конкретних індексів пакетів:
    
Якщо ви хочете встановити пакети з конкретного індексу пакетів (не з PyPI), ви можете вказати це в requirements.txt:
    
```
--index-url http://example.com/simple/
```

Залежності для конкретних платформ:
    
Ви можете вказати пакети, які потрібно встановити лише для певних платформ:
```
somepackage; sys_platform == "win32"
```

Використання зовнішніх вимог:

Ви можете включити залежності з іншого файлу requirements.txt:

```
-r other-requirements.txt
```

Встановлення пакетів без залежностей:
Якщо ви хочете встановити пакет без його залежностей, ви можете використовувати наступний синтаксис:
```
package_name --no-deps
```

### Безпека

При використанні pip важливо пам'ятати про безпеку. Не встановлюйте пакети з ненадійних джерел, і завжди перевіряйте код пакетів, якщо ви не впевнені в їхній безпеці.
