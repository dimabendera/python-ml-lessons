#### Частина 3: Пояснення коду

*   **`pygame.init()`** ініціалізує всі модулі Pygame, необхідні для роботи.
*   **`screen = pygame.display.set_mode(screen_size)`** створює вікно гри заданого розміру.
*   **`pygame.display.set_caption("Моя перша гра на Pygame")`** задає заголовок вікну гри.
*   Головний цикл гри (**`while True:`**) відповідає за обробку подій (наприклад, закриття вікна гри) і оновлення екрану.
*   **`pygame.event.get()`** отримує список всіх подій, що сталися з часу останнього виклику.
*   **`pygame.quit()`** завершує роботу Pygame, а **`sys.exit()`** — вихід із програми.

Цей приклад є основою для створення ігор на Pygame. Ви можете додавати графіку, звуки, обробку клавіатури та миші, щоб створити повноцінну гру.
