- Матричне множення автоматично застосує трансформацію до кожного пікселя.
- Якщо ви зіткнетеся з помилками форми (shape errors), перевірте розміри ваших масивів і переконайтеся, що вони сумісні для операцій.

Спробуйте реалізувати функцію `apply_sepia` самостійно.

`````python
import numpy as np
from PIL import Image
`````

`````python
# Функція для застосування фільтра сепія
def apply_sepia(image_array):
    # Коефіцієнти для перетворення RGB в сепію
    sepia_filter = np.array([[0.393, 0.769, 0.189],
                             [0.349, 0.686, 0.168],
                             [0.272, 0.534, 0.131]])
    
    # Перетворення зображення в сепію
    sepia_image = image_array @ sepia_filter.T
    sepia_image = np.clip(sepia_image, 0, 255)  # Обмеження значень від 0 до 255
    
    return sepia_image.astype(np.uint8)
`````

`````python
# Завантаження зображення за допомогою PIL
image = Image.open('sam.jpg')  # Вкажіть шлях до вашого зображення
image_array = np.array(image)

# Застосування фільтра сепія
sepia_image_array = apply_sepia(image_array)

# Створення нового зображення з отриманого масиву
sepia_image = Image.fromarray(sepia_image_array)

# Показ зображення
sepia_image
`````
